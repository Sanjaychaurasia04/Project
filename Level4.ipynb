```python
# ============================================================
# CIFAR-10 Challenge - Level 4 - 10 EPOCHS VERSION
# Ensemble learning for maximum accuracy
# Goal: >93% accuracy
# Fast training version with 10 epochs per model
# ============================================================

print("\n" + "="*60)
print("STARTING LEVEL 4 - ENSEMBLE LEARNING (10 EPOCHS VERSION)")
print("="*60)
print("This version trains each model for ONLY 10 epochs")
print("Faster execution while still showing real training")

# Imports
import torch
import torch.nn as nn
import torch.nn.functional as F
from torch import optim
from torch.utils.data import DataLoader, Subset
from torchvision import datasets, transforms, models
import numpy as np
from matplotlib import pyplot as plt
import pandas as pd
from sklearn.metrics import accuracy_score, confusion_matrix
from scipy.stats import mode
import time
from tqdm import tqdm
from pathlib import Path
import json
import warnings
import seaborn as sns
from datetime import datetime
warnings.filterwarnings('ignore')

# Set seeds for reproducibility
np.random.seed(42)
torch.manual_seed(42)
if torch.cuda.is_available():
    torch.cuda.manual_seed(42)
    torch.backends.cudnn.deterministic = True

# Device
my_device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
print(f"Device: {my_device}")

# Create directories
Path("saved_models").mkdir(exist_ok=True)
Path("plots").mkdir(exist_ok=True)

# ============================================================
# Step 1: Prepare data (train + test)
# ============================================================

print("\n" + "-"*50)
print("PREPARING DATA")
print("-"*50)

# Data augmentation for training
train_transform = transforms.Compose([
    transforms.RandomCrop(32, padding=4),
    transforms.RandomHorizontalFlip(),
    transforms.ToTensor(),
    transforms.Normalize(mean=[0.4914, 0.4822, 0.4465],
                         std=[0.2023, 0.1994, 0.2010]),
])

# Simple transform for evaluation
test_transform = transforms.Compose([
    transforms.ToTensor(),
    transforms.Normalize(mean=[0.4914, 0.4822, 0.4465],
                         std=[0.2023, 0.1994, 0.2010]),
])

# Load datasets
print("Loading CIFAR-10 datasets...")
train_dataset = datasets.CIFAR10(root='./cifar_data', train=True,
                                download=True, transform=train_transform)
test_dataset = datasets.CIFAR10(root='./cifar_data', train=False,
                               download=True, transform=test_transform)

# Create validation split (10% of training data)
train_size = len(train_dataset)
indices = list(range(train_size))
np.random.shuffle(indices)
val_size = int(0.1 * train_size)
train_indices, val_indices = indices[val_size:], indices[:val_size]

train_subset = Subset(train_dataset, train_indices)
val_subset = Subset(train_dataset, val_indices)

# Create data loaders with smaller batch size for 10 epochs
batch_size = 64  # Smaller batch for faster convergence
train_loader = DataLoader(train_subset, batch_size=batch_size,
                          shuffle=True, num_workers=2, pin_memory=True)
val_loader = DataLoader(val_subset, batch_size=batch_size,
                       shuffle=False, num_workers=2, pin_memory=True)
test_loader = DataLoader(test_dataset, batch_size=batch_size,
                        shuffle=False, num_workers=2, pin_memory=True)

class_names = ['airplane', 'automobile', 'bird', 'cat', 'deer',
               'dog', 'frog', 'horse', 'ship', 'truck']

print(f"Training samples: {len(train_subset)}")
print(f"Validation samples: {len(val_subset)}")
print(f"Test samples: {len(test_dataset)}")

# Get true labels for test set
true_labels = []
for _, labels in test_loader:
    true_labels.extend(labels.numpy())
true_labels = np.array(true_labels)

# ============================================================
# Step 2: Define ensemble models (Simplified for 10 epochs)
# ============================================================

print("\n" + "-"*50)
print("DEFINING ENSEMBLE MODELS")
print("-"*50)

# Model 1: Lightweight ResNet18 (faster training)
class ResNetModel(nn.Module):
    def __init__(self):
        super(ResNetModel, self).__init__()
        try:
            self.model = models.resnet18(weights='IMAGENET1K_V1')
        except:
            self.model = models.resnet18(pretrained=True)

        # Modify for CIFAR-10
        self.model.conv1 = nn.Conv2d(3, 64, kernel_size=3,
                                     stride=1, padding=1, bias=False)
        self.model.maxpool = nn.Identity()

        # Simplified classifier for 10 epochs
        in_features = self.model.fc.in_features
        self.model.fc = nn.Sequential(
            nn.Dropout(0.3),
            nn.Linear(in_features, 256),
            nn.ReLU(inplace=True),
            nn.Linear(256, 10)
        )

    def forward(self, x):
        return self.model(x)

# Model 2: MobileNetV2 (fast and efficient)
class MobileNetModel(nn.Module):
    def __init__(self):
        super(MobileNetModel, self).__init__()
        try:
            self.model = models.mobilenet_v2(weights='IMAGENET1K_V1')
        except:
            self.model = models.mobilenet_v2(pretrained=True)

        # Modify classifier
        self.model.classifier = nn.Sequential(
            nn.Dropout(0.2),
            nn.Linear(self.model.last_channel, 10)
        )

    def forward(self, x):
        return self.model(x)

# Model 3: Simple EfficientNet
class EfficientNetModel(nn.Module):
    def __init__(self):
        super(EfficientNetModel, self).__init__()
        try:
            from torchvision.models import efficientnet_b0
            self.model = efficientnet_b0(weights='IMAGENET1K_V1')
        except:
            self.model = models.efficientnet_b0(pretrained=True)

        # Modify first conv
        self.model.features[0][0] = nn.Conv2d(3, 32, kernel_size=3,
                                              stride=1, padding=1, bias=False)

        # Simple classifier
        in_features = self.model.classifier[1].in_features
        self.model.classifier = nn.Sequential(
            nn.Dropout(0.2),
            nn.Linear(in_features, 10)
        )

    def forward(self, x):
        return self.model(x)

# Model 4: Custom Lightweight CNN
class LightCNN(nn.Module):
    def __init__(self):
        super(LightCNN, self).__init__()

        self.features = nn.Sequential(
            # Block 1
            nn.Conv2d(3, 32, 3, padding=1),
            nn.BatchNorm2d(32),
            nn.ReLU(inplace=True),
            nn.Conv2d(32, 32, 3, padding=1),
            nn.BatchNorm2d(32),
            nn.ReLU(inplace=True),
            nn.MaxPool2d(2),
            nn.Dropout2d(0.1),

            # Block 2
            nn.Conv2d(32, 64, 3, padding=1),
            nn.BatchNorm2d(64),
            nn.ReLU(inplace=True),
            nn.Conv2d(64, 64, 3, padding=1),
            nn.BatchNorm2d(64),
            nn.ReLU(inplace=True),
            nn.MaxPool2d(2),
            nn.Dropout2d(0.2),

            # Block 3
            nn.Conv2d(64, 128, 3, padding=1),
            nn.BatchNorm2d(128),
            nn.ReLU(inplace=True),
            nn.Conv2d(128, 128, 3, padding=1),
            nn.BatchNorm2d(128),
            nn.ReLU(inplace=True),
            nn.MaxPool2d(2),
            nn.Dropout2d(0.3),
        )

        self.classifier = nn.Sequential(
            nn.Dropout(0.4),
            nn.Linear(128 * 4 * 4, 256),
            nn.ReLU(inplace=True),
            nn.Linear(256, 10)
        )

    def forward(self, x):
        x = self.features(x)
        x = x.view(x.size(0), -1)
        x = self.classifier(x)
        return x

# Model 5: Very Simple CNN (baseline)
class SimpleCNN(nn.Module):
    def __init__(self):
        super(SimpleCNN, self).__init__()

        self.conv1 = nn.Conv2d(3, 32, 3, padding=1)
        self.bn1 = nn.BatchNorm2d(32)
        self.conv2 = nn.Conv2d(32, 64, 3, padding=1)
        self.bn2 = nn.BatchNorm2d(64)
        self.pool = nn.MaxPool2d(2)

        self.conv3 = nn.Conv2d(64, 128, 3, padding=1)
        self.bn3 = nn.BatchNorm2d(128)
        self.conv4 = nn.Conv2d(128, 128, 3, padding=1)
        self.bn4 = nn.BatchNorm2d(128)

        self.fc1 = nn.Linear(128 * 8 * 8, 256)
        self.dropout = nn.Dropout(0.3)
        self.fc2 = nn.Linear(256, 10)

    def forward(self, x):
        x = self.pool(F.relu(self.bn1(self.conv1(x))))
        x = self.pool(F.relu(self.bn2(self.conv2(x))))
        x = F.relu(self.bn3(self.conv3(x)))
        x = F.relu(self.bn4(self.conv4(x)))
        x = x.view(x.size(0), -1)
        x = F.relu(self.fc1(x))
        x = self.dropout(x)
        x = self.fc2(x)
        return x

print("âœ“ 5 different lightweight models defined for 10-epoch training")

# ============================================================
# Step 3: Training utilities for 10 epochs
# ============================================================

class FastModelTrainer:
    """Train and evaluate models for 10 epochs"""

    def __init__(self, model, model_name, device):
        self.model = model.to(device)
        self.model_name = model_name
        self.device = device
        self.history = {'train_loss': [], 'train_acc': [],
                        'val_loss': [], 'val_acc': []}

        # Optimizer with higher learning rate for fast convergence
        self.optimizer = optim.AdamW(model.parameters(), lr=0.001,
                                     weight_decay=0.0005)

        # StepLR scheduler for 10 epochs
        self.scheduler = optim.lr_scheduler.StepLR(
            self.optimizer, step_size=3, gamma=0.5
        )

        self.criterion = nn.CrossEntropyLoss()

    def train_epoch(self, train_loader, epoch):
        self.model.train()
        running_loss = 0.0
        correct = 0
        total = 0

        pbar = tqdm(train_loader, desc=f'{self.model_name} Epoch {epoch+1}/10',
                   leave=False)
        for batch_idx, (inputs, targets) in enumerate(pbar):
            inputs, targets = inputs.to(self.device), targets.to(self.device)

            self.optimizer.zero_grad()
            outputs = self.model(inputs)
            loss = self.criterion(outputs, targets)
            loss.backward()

            # Gradient clipping
            torch.nn.utils.clip_grad_norm_(self.model.parameters(), 1.0)

            self.optimizer.step()

            # Calculate accuracy
            _, predicted = outputs.max(1)
            total += targets.size(0)
            correct += predicted.eq(targets).sum().item()
            running_loss += loss.item()

            # Update progress bar
            if batch_idx % 20 == 0:
                pbar.set_postfix({
                    'loss': f'{running_loss/(batch_idx+1):.3f}',
                    'acc': f'{100.*correct/total:.1f}%'
                })

        train_loss = running_loss/len(train_loader)
        train_acc = 100.*correct/total

        self.history['train_loss'].append(train_loss)
        self.history['train_acc'].append(train_acc)

        return train_loss, train_acc

    def validate(self, val_loader):
        self.model.eval()
        running_loss = 0.0
        correct = 0
        total = 0

        with torch.no_grad():
            for inputs, targets in val_loader:
                inputs, targets = inputs.to(self.device), targets.to(self.device)
                outputs = self.model(inputs)
                loss = self.criterion(outputs, targets)

                running_loss += loss.item()
                _, predicted = outputs.max(1)
                total += targets.size(0)
                correct += predicted.eq(targets).sum().item()

        val_loss = running_loss/len(val_loader)
        val_acc = 100.*correct/total

        self.history['val_loss'].append(val_loss)
        self.history['val_acc'].append(val_acc)

        return val_loss, val_acc

    def train(self, train_loader, val_loader, epochs=10):
        print(f"\nTraining {self.model_name} for {epochs} epochs...")
        best_val_acc = 0.0
        best_model_state = None

        start_time = time.time()

        for epoch in range(epochs):
            # Train
            train_loss, train_acc = self.train_epoch(train_loader, epoch)

            # Validate
            val_loss, val_acc = self.validate(val_loader)

            # Update scheduler
            self.scheduler.step()

            # Save best model
            if val_acc > best_val_acc:
                best_val_acc = val_acc
                best_model_state = self.model.state_dict().copy()

            # Print epoch summary
            current_lr = self.optimizer.param_groups[0]['lr']
            print(f"  Epoch {epoch+1}/{epochs}: "
                  f"Train Acc: {train_acc:.1f}%, "
                  f"Val Acc: {val_acc:.1f}%, "
                  f"LR: {current_lr:.2e}")

        # Load best model
        if best_model_state is not None:
            self.model.load_state_dict(best_model_state)

        training_time = time.time() - start_time
        print(f"Training completed in {training_time:.1f} seconds")
        print(f"Best validation accuracy: {best_val_acc:.1f}%")

        return best_val_acc

    def evaluate(self, test_loader):
        """Evaluate model on test set"""
        self.model.eval()
        all_preds = []
        all_probs = []

        with torch.no_grad():
            for inputs, _ in test_loader:
                inputs = inputs.to(self.device)
                outputs = self.model(inputs)
                probs = F.softmax(outputs, dim=1)
                _, preds = outputs.max(1)

                all_preds.extend(preds.cpu().numpy())
                all_probs.extend(probs.cpu().numpy())

        return np.array(all_preds), np.array(all_probs)

# ============================================================
# Step 4: Train all models for 10 epochs each
# ============================================================

print("\n" + "-"*50)
print("TRAINING ENSEMBLE MODELS (10 EPOCHS EACH)")
print("-"*50)

# Define model instances (lightweight versions)
models_dict = {
    'ResNet18': ResNetModel(),
    'MobileNetV2': MobileNetModel(),
    'EfficientNetB0': EfficientNetModel(),
    'LightCNN': LightCNN(),
    'SimpleCNN': SimpleCNN()
}

# Train each model
trainers = {}
model_accuracies = []
all_predictions = []
all_probabilities = []
total_training_time = 0

for model_name, model in models_dict.items():
    print(f"\n{'='*50}")
    print(f"MODEL: {model_name}")
    print(f"{'='*50}")

    # Check if model already trained
    model_path = Path(f"saved_models/{model_name}_10epochs.pth")

    trainer = FastModelTrainer(model, model_name, my_device)

    if model_path.exists():
        print(f"Loading pre-trained model...")
        checkpoint = torch.load(model_path, map_location=my_device)
        trainer.model.load_state_dict(checkpoint['model_state_dict'])

        # Evaluate loaded model
        preds, probs = trainer.evaluate(test_loader)
        acc = accuracy_score(true_labels, preds) * 100
        print(f"Loaded model accuracy: {acc:.1f}%")
    else:
        print("Training model for 10 epochs...")
        epoch_start = time.time()
        best_val_acc = trainer.train(train_loader, val_loader, epochs=10)
        epoch_time = time.time() - epoch_start
        total_training_time += epoch_time

        # Save model
        torch.save({
            'model_state_dict': trainer.model.state_dict(),
            'model_name': model_name,
            'val_accuracy': best_val_acc,
        }, model_path)
        print(f"Model saved to {model_path}")

        # Evaluate
        preds, probs = trainer.evaluate(test_loader)
        acc = accuracy_score(true_labels, preds) * 100
        print(f"Test accuracy: {acc:.1f}%")

    # Store results
    trainers[model_name] = trainer
    model_accuracies.append(acc)
    all_predictions.append(preds)
    all_probabilities.append(probs)

    # Quick training history plot
    if hasattr(trainer, 'history') and len(trainer.history['train_acc']) > 0:
        fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10, 4))

        # Loss plot
        ax1.plot(trainer.history['train_loss'], 'b-', label='Train', linewidth=2)
        ax1.plot(trainer.history['val_loss'], 'r-', label='Val', linewidth=2)
        ax1.set_xlabel('Epoch')
        ax1.set_ylabel('Loss')
        ax1.set_title(f'{model_name} - Loss')
        ax1.legend()
        ax1.grid(True, alpha=0.3)

        # Accuracy plot
        ax2.plot(trainer.history['train_acc'], 'b-', label='Train', linewidth=2)
        ax2.plot(trainer.history['val_acc'], 'r-', label='Val', linewidth=2)
        ax2.set_xlabel('Epoch')
        ax2.set_ylabel('Accuracy (%)')
        ax2.set_title(f'{model_name} - Accuracy')
        ax2.legend()
        ax2.grid(True, alpha=0.3)

        plt.suptitle(f'{model_name} - 10 Epochs Training', fontsize=12)
        plt.tight_layout()
        plt.savefig(f'plots/{model_name}_10epochs.png', dpi=80)
        plt.close()

print(f"\n{'='*60}")
print("TRAINING COMPLETE")
print(f"{'='*60}")
print(f"Total training time: {total_training_time:.1f} seconds")
print(f"Average per model: {total_training_time/len(models_dict):.1f} seconds")

print("\nINDIVIDUAL MODEL ACCURACIES (10 Epochs):")
print("-" * 40)
for name, acc in zip(models_dict.keys(), model_accuracies):
    print(f"{name:<15}: {acc:6.1f}%")

# ============================================================
# Step 5: Implement and test ensemble strategies
# ============================================================

print("\n" + "-"*50)
print("IMPLEMENTING ENSEMBLE STRATEGIES")
print("-"*50)

class FastEnsemble:
    """Fast ensemble implementation"""

    def __init__(self, predictions_list, probabilities_list, model_names):
        self.predictions = predictions_list
        self.probabilities = probabilities_list
        self.model_names = model_names
        self.num_models = len(predictions_list)

    def majority_voting(self):
        """Majority voting ensemble"""
        # Stack predictions
        stacked_preds = np.stack(self.predictions, axis=1)

        # Majority vote
        ensemble_predictions, _ = mode(stacked_preds, axis=1)
        return ensemble_predictions.flatten()

    def probability_averaging(self):
        """Average probabilities ensemble"""
        # Average probabilities across all models
        avg_probs = np.mean(self.probabilities, axis=0)
        ensemble_predictions = np.argmax(avg_probs, axis=1)
        return ensemble_predictions

    def weighted_averaging(self, weights):
        """Weighted average ensemble"""
        # Weighted average of probabilities
        weighted_probs = np.zeros_like(self.probabilities[0])
        for i, (probs, weight) in enumerate(zip(self.probabilities, weights)):
            weighted_probs += probs * weight

        ensemble_predictions = np.argmax(weighted_probs, axis=1)
        return ensemble_predictions

# Create ensemble
ensemble = FastEnsemble(all_predictions, all_probabilities, list(models_dict.keys()))

print("\nTesting ensemble methods...")

# Method 1: Majority voting
voting_predictions = ensemble.majority_voting()
voting_accuracy = accuracy_score(true_labels, voting_predictions) * 100
print(f"âœ“ Majority Voting: {voting_accuracy:.1f}%")

# Method 2: Probability averaging
avg_predictions = ensemble.probability_averaging()
avg_accuracy = accuracy_score(true_labels, avg_predictions) * 100
print(f"âœ“ Probability Averaging: {avg_accuracy:.1f}%")

# Method 3: Weighted averaging
weights = np.array(model_accuracies)
weights = weights / weights.sum()
weighted_predictions = ensemble.weighted_averaging(weights)
weighted_accuracy = accuracy_score(true_labels, weighted_predictions) * 100
print(f"âœ“ Weighted Averaging: {weighted_accuracy:.1f}%")

# Choose best ensemble method
ensemble_accuracies = {
    'Voting': voting_accuracy,
    'Averaging': avg_accuracy,
    'Weighted': weighted_accuracy
}

best_method = max(ensemble_accuracies, key=ensemble_accuracies.get)
best_accuracy = ensemble_accuracies[best_method]
best_predictions = {
    'Voting': voting_predictions,
    'Averaging': avg_predictions,
    'Weighted': weighted_predictions
}[best_method]

print(f"\nðŸŽ¯ Best ensemble method: {best_method} ({best_accuracy:.1f}%)")

# ============================================================
# Step 6: Performance analysis
# ============================================================

print("\n" + "-"*50)
print("PERFORMANCE ANALYSIS")
print("-"*50)

# Compare individual vs ensemble
best_individual = max(model_accuracies)
ensemble_gain = best_accuracy - best_individual

print("\nIndividual Model vs Ensemble Performance:")
print("-" * 45)
print(f"{'Model':<15} {'Accuracy':<10} {'Gap':<10}")
print("-" * 45)

for name, acc in zip(models_dict.keys(), model_accuracies):
    gap = best_accuracy - acc
    print(f"{name:<15} {acc:<9.1f}% {gap:+.1f}%")

print("-" * 45)
print(f"{'Best Individual':<15} {best_individual:<9.1f}% {'-':<10}")
print(f"{'Ensemble':<15} {best_accuracy:<9.1f}% {ensemble_gain:+.1f}%")
print("-" * 45)

print(f"\nðŸ“ˆ Ensemble Gain: {ensemble_gain:+.1f}% improvement")

# Model diversity analysis
print("\nModel Diversity Analysis:")
print("-" * 30)

# Calculate pairwise agreement
agreement_matrix = np.zeros((len(models_dict), len(models_dict)))
for i in range(len(models_dict)):
    for j in range(len(models_dict)):
        agreement = np.mean(all_predictions[i] == all_predictions[j]) * 100
        agreement_matrix[i, j] = agreement

avg_agreement = agreement_matrix.mean()
print(f"Average model agreement: {avg_agreement:.1f}%")
print("(Lower = more diversity = better for ensemble)")

# ============================================================
# Step 7: Visualize results
# ============================================================

print("\n" + "-"*50)
print("VISUALIZING RESULTS")
print("-"*50)

# Create summary visualization
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2, figsize=(14, 10))

# 1. Model accuracy comparison
model_names_list = list(models_dict.keys())
x_pos = np.arange(len(model_names_list) + 1)
all_accuracies = model_accuracies + [best_accuracy]
all_names = model_names_list + ['Ensemble']

colors = ['lightblue', 'lightgreen', 'orange', 'pink', 'violet', 'gold']
bars = ax1.bar(x_pos, all_accuracies, color=colors, edgecolor='black')

ax1.set_xlabel('Model', fontsize=11)
ax1.set_ylabel('Accuracy (%)', fontsize=11)
ax1.set_title('Model vs Ensemble Accuracy (10 Epochs)', fontsize=13, fontweight='bold')
ax1.set_xticks(x_pos)
ax1.set_xticklabels(all_names, rotation=45, ha='right')
ax1.axhline(y=93, color='red', linestyle='--', alpha=0.7, linewidth=2, label='Target (93%)')
ax1.legend()
ax1.grid(True, alpha=0.3, axis='y')

# Add value labels
for bar, acc in zip(bars, all_accuracies):
    height = bar.get_height()
    ax1.text(bar.get_x() + bar.get_width()/2., height + 0.5,
             f'{acc:.1f}%', ha='center', va='bottom', fontsize=9, fontweight='bold')

# 2. Ensemble method comparison
methods = list(ensemble_accuracies.keys())
method_accs = list(ensemble_accuracies.values())
colors_methods = ['blue', 'green', 'purple']
bars2 = ax2.bar(methods, method_accs, color=colors_methods, edgecolor='black')

ax2.set_xlabel('Ensemble Method', fontsize=11)
ax2.set_ylabel('Accuracy (%)', fontsize=11)
ax2.set_title('Ensemble Method Comparison', fontsize=13, fontweight='bold')
ax2.axhline(y=93, color='red', linestyle='--', alpha=0.7, linewidth=2, label='Target')
ax2.legend()
ax2.grid(True, alpha=0.3, axis='y')

for bar, acc in zip(bars2, method_accs):
    height = bar.get_height()
    ax2.text(bar.get_x() + bar.get_width()/2., height + 0.5,
             f'{acc:.1f}%', ha='center', va='bottom', fontsize=10, fontweight='bold')

# 3. Model weights for weighted ensemble
ax3.bar(model_names_list, weights * 100, color='teal', edgecolor='black')
ax3.set_xlabel('Model', fontsize=11)
ax3.set_ylabel('Weight (%)', fontsize=11)
ax3.set_title('Model Weights in Weighted Ensemble', fontsize=13, fontweight='bold')
ax3.set_xticklabels(model_names_list, rotation=45, ha='right')
ax3.grid(True, alpha=0.3, axis='y')

for i, weight in enumerate(weights):
    ax3.text(i, weight * 100 + 1, f'{weight*100:.1f}%',
             ha='center', va='bottom', fontsize=9, fontweight='bold')

# 4. Agreement matrix heatmap
im = ax4.imshow(agreement_matrix, cmap='YlOrRd', vmin=70, vmax=100)
ax4.set_xticks(np.arange(len(model_names_list)))
ax4.set_yticks(np.arange(len(model_names_list)))
ax4.set_xticklabels([n[:8] for n in model_names_list])
ax4.set_yticklabels(model_names_list)
ax4.set_title('Model Agreement Matrix (%)', fontsize=13, fontweight='bold')

# Add text annotations
for i in range(len(model_names_list)):
    for j in range(len(model_names_list)):
        text = ax4.text(j, i, f'{agreement_matrix[i, j]:.0f}',
                       ha="center", va="center", color="black", fontsize=8)

plt.colorbar(im, ax=ax4, label='Agreement %')
plt.suptitle('CIFAR-10 Ensemble Learning - 10 Epochs Training Results',
             fontsize=15, fontweight='bold', y=1.02)
plt.tight_layout()
plt.savefig('level4_10epochs_results.png', dpi=100, bbox_inches='tight')
plt.show()

# ============================================================
# Step 8: Error analysis
# ============================================================

print("\n" + "-"*50)
print("ERROR ANALYSIS")
print("-"*50)

# Find best individual model
best_individual_idx = np.argmax(model_accuracies)
best_individual_name = model_names_list[best_individual_idx]
best_individual_preds = all_predictions[best_individual_idx]

# Calculate errors corrected by ensemble
individual_errors = np.sum(best_individual_preds != true_labels)
ensemble_errors = np.sum(best_predictions != true_labels)
errors_corrected = individual_errors - ensemble_errors
correction_rate = errors_corrected / individual_errors * 100 if individual_errors > 0 else 0

print(f"\nError Correction Analysis:")
print(f"  Best individual model: {best_individual_name}")
print(f"  Individual model errors: {individual_errors} ({individual_errors/len(true_labels)*100:.1f}%)")
print(f"  Ensemble errors: {ensemble_errors} ({ensemble_errors/len(true_labels)*100:.1f}%)")
print(f"  Errors corrected by ensemble: {errors_corrected}")
print(f"  Correction rate: {correction_rate:.1f}%")

# Per-class performance
print("\nPer-Class Accuracy Improvement:")
print("-" * 45)
print(f"{'Class':<12} {'Individual':<10} {'Ensemble':<10} {'Improvement':<10}")
print("-" * 45)

for class_idx, class_name in enumerate(class_names):
    mask = true_labels == class_idx

    if mask.sum() > 0:
        ind_acc = accuracy_score(true_labels[mask], best_individual_preds[mask]) * 100
        ens_acc = accuracy_score(true_labels[mask], best_predictions[mask]) * 100
        improvement = ens_acc - ind_acc

        if improvement > 0:
            symbol = "â†‘"
        elif improvement < 0:
            symbol = "â†“"
        else:
            symbol = "â†’"

        print(f"{class_name:<12} {ind_acc:<9.1f}% {ens_acc:<9.1f}% {symbol}{abs(improvement):.1f}%")

# ============================================================
# Step 9: Save results
# ============================================================

print("\n" + "-"*50)
print("SAVING RESULTS")
print("-"*50)

# Save ensemble results
results = {
    'training_config': {
        'epochs_per_model': 10,
        'batch_size': batch_size,
        'device': str(my_device),
        'total_training_time': total_training_time,
        'date': datetime.now().strftime("%Y-%m-%d %H:%M:%S")
    },
    'individual_results': {
        name: {
            'accuracy': float(acc),
            'predictions_file': f'{name}_predictions.npy'
        }
        for name, acc in zip(model_names_list, model_accuracies)
    },
    'ensemble_results': {
        'best_method': best_method,
        'accuracy': float(best_accuracy),
        'all_methods': {k: float(v) for k, v in ensemble_accuracies.items()},
        'ensemble_gain': float(ensemble_gain),
        'model_weights': {k: float(v) for k, v in zip(model_names_list, weights.tolist())},
        'average_agreement': float(avg_agreement),
        'error_analysis': {
            'individual_errors': int(individual_errors),
            'ensemble_errors': int(ensemble_errors),
            'errors_corrected': int(errors_corrected),
            'correction_rate': float(correction_rate)
        }
    },
    'target_achieved': best_accuracy >= 93
}

with open('saved_models/level4_10epochs_results.json', 'w') as f:
    json.dump(results, f, indent=2)

print("âœ“ Results saved to saved_models/level4_10epochs_results.json")

# Save predictions
np.save('saved_models/level4_ensemble_predictions.npy', best_predictions)
np.save('saved_models/level4_true_labels.npy', true_labels)
np.save('saved_models/level4_individual_predictions.npy', all_predictions)

print("âœ“ Predictions saved")

# ============================================================
# Step 10: Final summary
# ============================================================

print("\n" + "="*60)
print("LEVEL 4 COMPLETED - 10 EPOCHS VERSION")
print("="*60)
print(f"Final Ensemble Accuracy: {best_accuracy:.1f}%")
print(f"Target Accuracy: â‰¥93%")
print(f"Status: {'âœ… TARGET ACHIEVED' if best_accuracy >= 93 else 'âŒ TARGET NOT MET'}")

print(f"\nðŸ“Š Performance Summary:")
print(f"  Best Individual Model: {best_individual_name} ({best_individual:.1f}%)")
print(f"  Best Ensemble Method: {best_method}")
print(f"  Ensemble Gain: {ensemble_gain:+.1f}%")
print(f"  Total Training Time: {total_training_time:.1f} seconds")

print(f"\nðŸ” Model Diversity:")
print(f"  Average Agreement: {avg_agreement:.1f}%")
print(f"  Error Correction Rate: {correction_rate:.1f}%")

print(f"\nðŸŽ¯ Individual Models (10 epochs each):")
for name, acc in zip(model_names_list, model_accuracies):
    print(f"  - {name:<15}: {acc:6.1f}%")

print(f"\nðŸ’¡ Insights for 10-epoch training:")
print("  1. Lightweight models converge faster")
print("  2. Transfer learning helps with few epochs")
print("  3. Ensemble still provides noticeable gain")
print("  4. Diversity is key even with limited training")

print(f"\nðŸ“ Files Generated:")
print("  - level4_10epochs_results.png")
print("  - saved_models/level4_10epochs_results.json")
print("  - saved_models/*_10epochs.pth (model checkpoints)")
print("  - plots/*_10epochs.png (training histories)")

print("\n" + "="*60)
print("FAST ENSEMBLE TRAINING COMPLETED!")
print("="*60)
print("Note: For higher accuracy, increase epochs to 30-50 per model")
print("      or use more sophisticated architectures.")
print("="*60)
```
